# Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
x=np.arange(0,10)
y=np.arange(10,20)
```

# To use scatter plot 

```python
plt.scatter(x,y,c='b')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('scatter plot')
#To save graph
plt.savefig('fig.jpeg')
```

# To use plt.plot

```python
plt.plot(x,y,'g*--')
```

```python
y=x*x
```

```python
plt.plot(x,y,'bo--')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('plot with value y=x*x')
```

# To use subplot

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r*--')
plt.subplot(2,2,2)
plt.plot(x,y,'b.--')
plt.subplot(2,2,4)
plt.plot(x,y,'go--')
```

# To create sine wave

```python
np.pi
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('sine wave')
plt.plot(x,y)
```

# To create cos wave 

```python
np.pi
```

```python
x=np.arange(0,5*np.pi,0.1)
y=np.cos(x)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('cos wave')
plt.plot(x,y)
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.subplot(2,1,1)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('sine wave')
plt.plot(x,y_sin,'r--')
plt.subplot(2,1,2)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('cos wave')
plt.plot(x,y_cos,'g--')
plt.show()
```

# To use barplot

```python
x1=[2,3,4,5]
y1=[1,5,4,6]
plt.xlabel('x1 values')
plt.ylabel('y1-values')
plt.title('bar plot')
plt.bar(x1,y1)
```

```python
x1=[2,3,4,5]
y1=[1,5,4,6]
x2=[6,4,8,1]
y2=[9,11,2,4]
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('bar plot')
plt.bar(x1,y1)
plt.bar(x2,y2,color='g')
```

# To use Histogram

```python
df=np.array([54,2,34,12,87,27,52,5,98,53,99,13,22,25])
plt.title('histogram')
plt.hist(df)
```

```python
plt.title('histogram')
plt.hist(df,bins=20)
plt.show()
```

# To use boxplot

```python
df=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(df,vert=True,patch_artist=True)
plt.title('boxplot')
plt.show()
```

# To use pieplot

```python
labels='Numpy','Pandas','Matplotlib','Seaborn'
sizes=[220,240,120,190]
colors=['gold','green','red','blue']
explode=(0,0,0.1,0)
plt.pie(sizes, colors=colors, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True)
plt.title('pie plot')
plt.plot()
```
